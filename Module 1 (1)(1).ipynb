```python
lt["Day_Perc_Change"]= lt["Close Price"].pct_change()*100#Adding a new column Day_Perc_Change
```

The first element is nan because there is no previous element to calculate the percentage.

```python
lt["Day_Perc_Change"][0] = 0
```

**Answer 1.7**

```python
def trend(lt):
    prc = (np.nditer(lt["Day_Perc_Change"]))
    for i in prc:
        if i>=-0.5 and i<=0.5:
            return "Slight or No change"
        if i>0.5 and i<=1:
            return "Slight Positive"
        if i>=1 and i<3:
            return "Positive"
        if i>=3 and i<7:
            return "Among Top Gainers"
        if i<-0.5 and i>=-1:
            return "Slight Negative"
        if i<=-3 and i>-7:
            return "Among top losers"
        if i<-1 and i>-3:
            return "Negative"
        if i>=7:
            return "Bull run"
        if i<=-7:
            return "Bear drop"
            
```

```python
lt["Trend"] = lt.apply(trend,axis=1)
```

```python
lt["Trend"].value_counts()
```
